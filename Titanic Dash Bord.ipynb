```python
@app.callback(
            Output(component_id="g",component_property="figure"),
            Input(component_id='slider',component_property="value"),

)
def fun(input1):
    
    g1=px.scatter(df[(df['Pclass']==input1) ],x='Age',y='Fare',color='Sex',symbol='Survived',
           size='Pclass',size_max=10,log_x=True,
           template='plotly_dark')
    
 
    
    return g1
            
    
```
